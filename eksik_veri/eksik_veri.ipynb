## Eksik Verinin Önemi
* Eksik veri analizin doğrulunu etkiler
* Veri kalitesi ve güvenilirliğini düşürür
* Model performansını zayıflatır
* Karar verme süreçlerinde hatalara yol açabilir


## Eksik Verinin Oluşma Nedenleri
* Katılmcıların soruları atlaması (anket sorularında)
* Sensör veya sistem hataları (iot cihaz, hava kalitesi sensörü)
* Veri toplama sürecindeki eksilikler (hastane sisteminde personel hatası)
* Bilinçli olarak paylaşılmayan bilgiler
* Dosya veya aktarım hataları

## Eksik Veri Türleri
* MCAR (Missing Completely at Random): Eksiklik tamamen rastgeledir. İstemeden veri kaybetme.
* MAR (Missing at Random): Eksiklik başka bir değişkene bağlıdır. Yaş arttıkça maaş paylaşımı azalır gibi.
* MNAR (Missing Not at Random): Eksiklik kendi değişkeniyle ilgilidir.
* Türü belirlemek -> doğru müdahele seçimi için kritik

# Eksik Veriyi Tespit Etme Yöntemleri
* Gözle kontrol ve özetleme tabloları (excel boş hücreler, NaN değerler)
* Veri setinde boş hücrelerin sayımı
* Eksik oranlarının hesaplanması
* Görselleştirme (ısı haritası, sütun grafikleri)
* Eksik veri raporu oluşturma (Data Quality Report)



## Eksik Veriyle İlgili Karar Süreci
* Eksik verinin miktarı analiz edilir
* Eksikliğin rastgele olup olmadığı değerlendirilir
* Verinin önemi belirlenir.
* Silme veya doldurma kararı verilir.
* Gerekirse gelişmiş yöntemler uygulanır.

## Eksik Veri Doldurmanın Amacı
* Eksik değerleri koruyarak veri setini kullanılır hale getirmek
* Veri kaybını en aza indirmek
* Modelin öğrenme sürecini dengede tutmak
* Gerçek dağılıma en yakın değerlerle tamamlamak

## Ortalama (Mean) ile Doldurma
* Sayısal değişkenler için uygundur (yaş,gelir,sıcaklık gibi sürekli değerler)
* Basit ve hızlı bir yöntemdir
* Uç değerlerden (outlier) etkilenir
* Dağılımı yapay olarak daraltabilir


## Medyan ile Doldurma
* Aykırı (uç) değerlerden etkilenmez
* Merken eğilimi korur
* Özellikle çarpık (skewed) dağılımlar için uygundur.
* Sıralı (ordinal) değişkenlerde tercih edilir.

## Mod ile Doldurma
* Kategorik veriler için uygundur.
* En sık tekrar eden değerlerle doldurur
* Basit ama etkili bir yöntemdir
* Dağılımda tek bir kategori baskın hale gelebilir. Bu sıkıntıdır tek bir değer baskın hale gelir.

## Yöntem Seçiminde Dikkat Edilmesi Gerekenler
* Veri türü -> yöntem seçimini belriler.
* Eksiklik oranı yüksekse istatiksel doldurma yeterli olmayabilir. Gelişmiş yöntem veya uzmana danışmak gerekebilir
* Doldurma sonrası veri dağılımı kontrol edilmelidir
* Gerektiğinde gelişmiş yöntemlere geçilmelidir.

## Neden Gelişmiş Yöntemler
* Basit doldurma yöntemleri her zaman yeterli değildir
* Veri arasındaki ilişkiler göz ardı edilir
* Gelişmiş yöntemler tahmine dayalı yaklaşımlar kullanılır
* Gerçekçi ve dinamik değerleri üretebilir

## KNN (K-Nearsent Neighbors) ile Doldurma
* Eksik değeri tahmin ederken benzer gözlemleri kullanır
* "En yakın k komşu" prensibiyle çalışır
* Her eksik değer için ortalama tahmin yapılır
* Karmaşık ama güçlü bir yöntemdir


## Regresyon ile Doldurma

* Eksik değeri tahmin etmek için regresyon modeli kurulur
* Eksik sütun bağımlı değişken olarak alınır
* Diğer sütunlar bağımsız değişken olarak kullanılır
* İlişkileri matematiksel olarak tahmin eder



## Gelişmiş Doldurma Uygulamalarında Dikkat Edilmesi Gerekenler
* Model doğruluğunu çapraz doğrulama ile test etmek gerekir
* Doldurma sonrası veri dağılımı kontrol edilmelidir
* Aşırı uydurma(overfitting) riski vardır
* Eksik verinin nedenini anlamak hala en kritik adımdır
