## Öznitelik Çıkarma Nedir?
- Veriden anlamlı bilgilerin çıkarılması sürecidir.
- Ham veriyi makine öğrenmesi algoritmalarının anlayabileceği sayısal temsillere dönüştürür.
- Her öznitelik, verinin belirli bir yönünü veya özelliğini temsil eder.
- Görselde kenarlar, metinde kelime sıklıkları, seste frekanslar birer özniteliktir.

## Öznitelik Çıkarma Neden Önemlidir?
- Modelin öğrenme başarısını doğrudan etkiler.
- Karmaşık veriyi daha düşük boyutlu ve anlamlı hale getirir.
- Gereksiz bilgileri eleyerek hesaplama maliyetini azaltır.
- Doğru öznitelikler -> Daha doğru tahminler.

## Öznitelik Çıkarma Nerelerde Kullanılır?
- Görüntü işleme (object detection, yüz tanıma)
- Metin madenciliği ve doğal dil işleme
- Ses tanıma ve konuşma analizi
- Finansal zaman serisi tahminleri
- Biyomedikal veri analizi

## Metin İşlemede Öznitelik Çıkarma
- Amaç: Metinleri sayısal temsile dönüştürmek.
- Yöntemler:
  - Bag of Words (BoW)
  - TF-IDF (Term Frequency- Inverse Document Frequency)
  - Word2Vec, GloVe, BERT gibi embedding tabanlı temsiller
- Her kelime veya kelime grubu bir vektörle temsil edilir.


## Görüntü İşlemede Öznitelik Çıkarma
- Görüntüler piksel bazında karmaşık yapılardır.
- Temel öznitelikler:
  - kenar,köşe,doku,renk histogramı
- Yöntemler:
  - SIFT,SRUF,HOG gibi klasik teknikler
  - CNN tabanlı otomatik öznitelik öğrenimi
- Amaç: Görselin yapısını daha az boyutlu ama anlamlı şekilde temsil etmek.


## Öznitelik Mühendisinin Sorumlulukları
- Ham veriyi analiz etmek ve anlamlı özellikler tasarlamak
- Veri temizleme, dönüştürme ve ölçeklendirme işlemlerini yürütmek
- Uygun öznitelik çıkarma tekniklerini seçmek
- Modelin performansını artıracak yeni öznitelikler üretmek (feature extraction,creation)
- Aşırı veya gereksiz öznitelikleri tespit edip elemek
- Veri bilimi ve makine öğrenmesi ekipleriyle iş birliği yapmak
- Süreçleri belgelendirmek ve yeniden kullanılabilir pipeline'lar geliştirmek

## Metinlerden Öznitelik Çıkarma Nedir?
- Metin verileri ,bilgisayar için doğrudan işlenebilir değildir. (mutluyum = 0.84)
- Amaç: Metinleri sayısal temsillere dönüştürmek
- Her kelime, cümle veya belge; modelin anlayabileceği vektörlerle temsil edilir.
- Bu temsiller, sınıflandırma, duygu analiz, konu modelleme gibi görevlerde kullanılır.


## Bag of Words Nedir?
- Bag of Words(BoW), doğal dil işleme (NLP) ve metin madenciliğinde kullanılan temel bir metin temsili yöntemidir.
- BoW, metinlerdeki kelimeleri sayısal verilere dönüştürür ve metinlerin analizini sağlar.

## BoW Yönteminin İşleyişi
- Kelime Kümesi Oluşturma
- Kelime Frekansı Hesaplama
- Vektör Temsili


## TF-IDF
- TF-IDF, metin madenciliğinde ve bilgi erişiminde sıkça kullanılan bir özellik çıkarım yöntemidir.
- TF-IDF, kelimelerin belgeler içinde ne kadar önemli olduğunu belirlemek için kullanılır.

- Team Frequency (TF): Bir kelimenin bir belgede ne kadar sık geçtiğini ölçer.
- Inverse Document Frequency (IDF): Bir kelimenin tüm belgelerideki yaygınlığını ölçer. Bir kelimenin çok belgede geçiyorsa, o kelime çok fazla bilgi sağlamaz.

## N-Gram Modelleri Nedir?
- Bir dil modelinde kullanılan kelime veya karakter dizisinin uzunluğunu belirten bir terimdir.
- N-Gram modelleri, metinleri n kelimelik veya n karakterlik kısımlara bölerek analiz eder.

## N-Gram Modelleri Kullanım Alanları
- Metin Modelleme
- Metin Sınıflandırma
- Metin Üretimi
- Metin Benzerliği

## Word Embeddings
- Word Embedding(Kelime Gömme), doğal dil işleme ve makine öğreniminde kullanılan bir tekniktir.
- Kelimeleri, genellikle sürekli bir vektör uzayında anlamlı temsil edecek şekilde sayısal vektörlere dönüştürür.
- Bu temsiller, kelimeler arasındaki anlamsal ve dilbilgisel ilişkileri yakalamayı hedefler.

## Word Embedding Modelleri
- Word2Vec: Google tarafından geliştirilen, kelimeleri vektörlere dönüştüren ve bu vektörleri dildeki ilişkileri yakalayacak şekilde eğiten modeldir.
- GloVe (Global Vector for Word Representation): Standford Üniversitesi tarafından geliştirilen, kelime gömme temsillerini kelime ortaklıklarını yakalyacak şekilde hesaplayan bir modeldir.
- FastText: Facebook tarafından geliştirilen ve kelime gömme temsillerini kelime alt-birimlerini de dikkate alarak hesaplayan bir modeldir.
